```python
#calculating % of missing values
features_with_na=[features for features in data_train.columns if data_train[features].isnull().sum()>1]
for features in features_with_na:
  print(features,np.round(data_train[features].isnull().mean(),2),"% missing values")
```

**since there are many missing values we have to find relationship between missing values and sales price**

```python
for features in features_with_na:
  data=data_train.copy()
  #lets make a variable that indicates 1 is missing value 
  data[features]=np.where(data[features].isnull(),1,0)
  data.groupby(features)["SalePrice"].median().plot.bar()
  plt.title(features)
  plt.show()
```

```python
#Data Trasformation 
print ("Original: \n") 
print (data_train.Street.value_counts(), "\n")
```

```python
# Turn into one hot encoding 
data_train['enc_street'] = pd.get_dummies(data_train.Street, drop_first=True)
```

```python
# Encoded 
print ('Encoded: \n') 
print (data_train.enc_street.value_counts())
```

```python

condition_pivot = data_train.pivot_table(index='SaleCondition',
                                    values='SalePrice', aggfunc=np.median)
condition_pivot.plot(kind='bar', color='blue')
plt.xlabel('Sale Condition')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=0)
plt.show()
```

```python
def encode(x):
  if x == 'Partial':
    return 1 
  else:
    return 0
data_train['enc_condition'] = data_train.SaleCondition.apply(encode)
```

```python
condition_pivot = data_train.pivot_table(index='enc_condition', values='SalePrice', aggfunc=np.median)
condition_pivot.plot(kind='bar', color='blue')
plt.xlabel('Encoded Sale Condition')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=0)
plt.show()
```

```python
#Interpolation of data 
data = data_train.select_dtypes(include=[np.number]).interpolate().dropna()
```

```python
sum(data.isnull().sum() != 0)
```

```python
# Linear Model for the  train and test
y = np.log(data_train.SalePrice)
X = data.drop(['SalePrice', 'Id'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=.33)
```

```python
from sklearn import linear_model
from sklearn.ensemble import GradientBoostingRegressor
lr = GradientBoostingRegressor()
```

```python
model=lr.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
predictions
```

```python
print("R^2 is: ", model.score(x_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error
print ('RMSE is: \n', mean_squared_error(y_test, predictions))
```
